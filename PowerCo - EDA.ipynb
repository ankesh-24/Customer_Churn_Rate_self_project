Feature Engineering

```python
price_df['price_date'] = pd.to_datetime(price_df['price_date'], format = '%Y-%m-%d')
price_df
```

```python
price_df.info()
```

```python
price_df['price_date'].value_counts()
```

```python
# Group off-peak prices by companies and month
monthly_price_by_id = price_df.groupby(['id', 'price_date']).agg({'price_off_peak_var': 'mean', 'price_off_peak_fix': 'mean'}).reset_index()

# Get January and December prices
jan_prices = monthly_price_by_id.groupby('id').first().reset_index()
dec_prices = monthly_price_by_id.groupby('id').last().reset_index()

# Calculate the difference
diff_off = pd.merge(dec_prices.rename(columns={ 'price_off_peak_var': 'dec_1', 'price_off_peak_fix': 'dec_2'}), jan_prices.drop(columns='price_date'), on='id')
diff_off['offpeak_diff_dec_january_energy'] = diff_off['dec_1'] - diff_off['price_off_peak_var']
diff_off['offpeak_diff_dec_january_power'] = diff_off['dec_2'] - diff_off['price_off_peak_fix']
diff_off = diff_off[['id', 'offpeak_diff_dec_january_energy', 'offpeak_diff_dec_january_power']]
diff_off
```

```python
# Group peak prices by companies and month
monthly_price_by_id = price_df.groupby(['id', 'price_date']).agg({'price_peak_var': 'mean', 'price_peak_fix': 'mean'}).reset_index()

# Get January and December prices
jan_prices = monthly_price_by_id.groupby('id').first().reset_index()
dec_prices = monthly_price_by_id.groupby('id').last().reset_index()

# Calculate the difference
diff_peak = pd.merge(dec_prices.rename(columns={ 'price_peak_var': 'dec_1', 'price_peak_fix': 'dec_2'}), jan_prices.drop(columns='price_date'), on='id')
diff_peak['peak_diff_dec_january_energy'] = diff_peak['dec_1'] - diff_peak['price_peak_var']
diff_peak['peak_diff_dec_january_power'] = diff_peak['dec_2'] - diff_peak['price_peak_fix']
diff_peak = diff_peak[['id', 'peak_diff_dec_january_energy', 'peak_diff_dec_january_power']]
diff_peak
```

Merge diff_off and diff_peak

```python
price_diff = pd.merge(diff_peak, diff_off, on = 'id')
price_diff
```

Merge client_diff and price_diff

```python
client_diff = pd.merge(client_df, price_diff, on = 'id')
client_diff
```

Binarize Variance of Prices


```python
from sklearn.preprocessing import Binarizer

client_diff['peak_energy'] = Binarizer().fit_transform(client_diff['peak_diff_dec_january_energy'].values.reshape(-1,1))
client_diff['peak_power'] = Binarizer().fit_transform(client_diff['peak_diff_dec_january_power'].values.reshape(-1,1))

client_diff['off_energy'] = Binarizer().fit_transform(client_diff['offpeak_diff_dec_january_energy'].values.reshape(-1,1))
client_diff['off_power'] = Binarizer().fit_transform(client_diff['offpeak_diff_dec_january_power'].values.reshape(-1,1))

client_diff
    
```

Drop Object Columns

```python
client_diff = client_diff.select_dtypes(exclude=['object'])
client_diff
```

Heatmap

```python
corr = client_diff.corr()
sns.heatmap(corr)
```
